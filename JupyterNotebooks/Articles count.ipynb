```python
import pandas as pd
```

#### Load Data

```python
data = pd.read_csv('bias_mini.csv')
```

```python
data.head()
```

```python
articles = data['scraped_at']
```

```python
def count(content, current_counts) :
    words = content.split()
    count_a = current_counts['A']
    count_an = current_counts['An']
    count_the = current_counts['The']
    for word in words:
        if word.lower() == 'a' :
            count_a += 1
        if word.lower() == 'an' :
            count_an += 1
        if word.lower() == 'the' :
            count_the += 1
    return (count_a, count_an, count_the)
```

```python
current_counts = {'A':0, 'An': 0, 'The': 0}
for article in articles:
    current_counts['A'], current_counts['An'], current_counts['The'] = count(article,current_counts)
```
